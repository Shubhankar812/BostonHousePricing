```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_boston
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
boston = load_boston()
```

### **Preparing the Dataset**

```python
dataset = pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
dataset.head()
```

```python
dataset['Price'] = boston.target
```

```python
df = dataset
```

### **EDA**
The dataset contains float values and it is not good for linear regression, to convert it we have to use Correlation

```python
df.corr()
# if correlation_value == 0  ; 2 quantities are hardly related
#  if correlation_value closer to 1  ; 2 quantities are positively related
#  if correlation_value closer to 0  ; 2 quantities are negatively related

# Also if 2 quantities are strongly related to each other , that is df.corr() value is closer to 1 then we can drop one of the two quantities, this way we can avoid multicollinearity
```

```python
# Independent and dependent features

X = df.iloc[:,:-1]    # independent
y = df.iloc[:,-1]     # dependent
```

```python
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

### **Applying standardization**
We need to standardize the data coz. we are eventually going to use Gradient Descent for optimization that works calculating the global minima and hence we have to bring all the data elements in one scale.

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
X_test
```

```python
lr = LinearRegression()
```

```python
# Training the model
lr.fit(X_train,y_train)
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
# Prediction with test data
reg_pred = lr.predict(X_test)
```

```python
reg_pred
```

```python
plt.scatter(y_test,reg_pred)
```

```python
residuals = y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind="kde")
```

```python
print("Mean Absolute Error :",mean_absolute_error(y_test,reg_pred))
print("Mean Squared Error :",mean_squared_error(y_test,reg_pred))
```

### **Calculating the R2 Score**


Formula

R^2 = 1 - SSR/SST

R^2 = coefficient of determination SSR = sum of squares of residuals SST = total sum of squares

```python
score = r2_score(y_test,reg_pred)
```

```python
score
```

### **Adjusted R2 = 1  [(1-R2)*(n-1)/(n-k-1)]**

where:

R2: The R2 of the model n: The number of observations k: The number of predictor variables

```python
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
